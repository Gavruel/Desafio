```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carregar a base de dados de histórico de vendas
dados = pd.read_excel("Dados.xlsx")
```

```python
# Converter a coluna "Data" em formato de data
dados['Data'] = pd.to_datetime(dados['Data'])
```

```python
# Definir a coluna "Data" como índice do dataframe
dados.set_index("Data", inplace=True)
```

```python
# Plotar a série temporal de vendas
plt.plot(dados)
plt.xlabel("Data")
plt.ylabel("Vendas")
plt.title("Série Temporal de Vendas")
plt.show()
```

```python
# Calcular a média móvel com um espaço de 7 dias
dados_rolling = dados.rolling(window=7).mean()
```

```python
# Plotar a média móvel
plt.plot(dados_rolling)
plt.xlabel("Data")
plt.ylabel("Vendas (Média Móvel)")
plt.title("Previsão de Demanda com Média Móvel")
plt.show()
```

```python
# Prever a demanda para os próximos 5 dias
ultimos_valores = dados_rolling.tail(1)["Vendas"].values[0]
previsao = [ultimos_valores for i in range(5)]
```

```python
# Plotar a previsão
data_futura = pd.date_range(start=dados.index[-1], periods=5, freq='D')
previsao = pd.Series(previsao, index=data_futura)
plt.plot(dados, label="Vendas Reais")
plt.plot(previsao, label="Previsão de Demanda")
plt.xlabel("Data")
plt.ylabel("Vendas")
plt.title("Previsão de Demanda com Média Móvel")
plt.legend()
plt.show()
```
